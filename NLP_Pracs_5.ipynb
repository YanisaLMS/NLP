```python
!pip install nltk
!pip install stanza

import nltk
import stanza

# Download the Hindi language model for Stanza
stanza.download('hi')

# Initialize the Hindi pipeline
nlp = stanza.Pipeline('hi')

def pos_tag_hindi(text):
  """
  Performs POS tagging for Hindi text using Stanza.

  Args:
    text: The Hindi text to be tagged.

  Returns:
    A list of tuples, where each tuple contains a word and its POS tag.
  """
  doc = nlp(text)
  tagged_words = []
  for sentence in doc.sentences:
    for word in sentence.words:
      tagged_words.append((word.text, word.upos))
  return tagged_words

# Example Hindi text
hindi_text = """
यह एक बड़ा उदाहरण है हिंदी भाषा के लिए।
यह वाक्य  प्रदर्शित करता है कि हम कैसे हिंदी पाठ का उपयोग कर सकते हैं।
"""

# Perform POS tagging
tagged_words = pos_tag_hindi(hindi_text)

# Print the tagged words
for word, tag in tagged_words:
  print(f"{word}: {tag}")

```
